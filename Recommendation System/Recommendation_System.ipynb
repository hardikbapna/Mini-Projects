```
tfidf = TfidfVectorizer()
tf_mat = tfidf.fit_transform(data["genre"])
```

```
cosine_dissim_mat = pairwise_distances(tf_mat, tf_mat, metric = "cosine")
```

```
cosine_sim_mat = 1 - pd.DataFrame(cosine_dissim_mat)
```

```
pd.DataFrame(cosine_sim_mat)
```

```
#Index number to Anime name
#For Converting Index number to corresponding Anime Name.

index_to_name = dict()
for row in data.iterrows():
  index_to_name[row[0]] = row[1][1]
```

```
#Anime name to index number
#For Converting Anime Name to corresponding Index Number.

name_to_index = dict()
for row in data.iterrows():
  name_to_index[row[1][1]] = row[0]
```

```
cosine_simmat_df = pd.DataFrame(cosine_sim_mat)
```

```
def Recommendation(index, N):
  
  #Sorting
  cos_tuple = [(k,v) for k,v in dict(cosine_simmat_df[index]).items()]
  cosine_sim_sorted = sorted(cos_tuple, key = lambda i: i[1], reverse = True)

  #TopN
  list_a = []
  for i in cosine_sim_sorted[0:N]:
      list_a.append(i[0])
        
  #Printing Recommendations
  list_name = [index_to_name[i] for i in list_a]
  print("Recommended Movie list: \n")
  print(list_name)
```

```
#Taking input of Movie name and 
#number of Recommendations required.
#Calling Recommendation function.

Movie = str(input("Enter the name of Anime:"))
TopN = int(input("Enter how many Recommendations you want:\n"))

Recommendation(name_to_index[Movie], TopN)
```
